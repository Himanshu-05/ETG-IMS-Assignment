# Importing packages and loading JSON file into a dictionary

```python
import json
import datetime
import pandas as pd
from IPython.display import display
```

```python
fr = open("record.json",'r')
r = fr.read()
fr.close()
record = json.loads(r)

sales={}
bill={}
```

# Function to display the catalog of products available

```python
def display_catalog():
    df=pd.DataFrame(eval(r))
    res=df.transpose()
    res.index.name="Product ID"
    display(res)
```

# Helper Function to purchase product/s

```python
def purchase_prod():
    while True:
        product_id = input("Enter unique product id of the medicine to puchase: (Type \'exit\' to exit) ")
        if product_id=='exit':
            break
        elif product_id in record.keys():
            print("The number of strips available for ",record[product_id]['name']," is ",record[product_id]['Quantity']," at Rs. ",record[product_id]['Price']," per strip")
            quant=int(input("How many strips of the medicine do you wish to buy ? "))
            if quant<=record[product_id]['Quantity']:
                record[product_id]['Quantity']-=quant
                print("Purchase successful!!")
                
                curr_time=str(datetime.datetime.now())
                sales[curr_time] = {'product ID': product_id,'name': record[product_id]['name'], 'MFG': record[product_id]['MFG'], 'EXP': record[product_id]['EXP'], 'Price': record[product_id]['Price'], 'Quantity': quant}
                js = json.dumps(sales)
                fd = open("sales.json",'w')
                fd.write(js)
                fd.close()
                
                if int(product_id) in bill.keys():
                    bill[int(product_id)]['Quantity']+=quant
                else:
                    bill[int(product_id)] = {'Name': record[product_id]['name'], 'MFG': record[product_id]['MFG'], 'EXP': record[product_id]['EXP'], 'Price': record[product_id]['Price'], 'Quantity': quant} 
                
                if record[product_id]['Quantity']==0:
                    del record[product_id]
                js = json.dumps(record)
                fd = open("record.json",'w')
                fd.write(js)
                fd.close()
                break
            elif quant>record[product_id]['Quantity']:
                print("The entered amount was greater than the available quantity of the medicine so unable to process.")
                choice=input("Would you like to retry ? (Y for retry, N to exit purchase portal) ")
                if choice=='Y':
                    continue
                else:
                    break
        else:
            print("The medicine you're trying to purchase does not exist in the inventory!")
            choice=input("Would you like to retry? (Y for retry, N to exit purchase portal) ")
            if choice=='Y':
                continue
            else:
                break
```

# Helper function to generate bill for the customer

```python
def generate_bill():
    if not bool(bill):
        print("NO MEDICINE PURCHASED YET!!")
    else:
        df=pd.DataFrame.from_dict(bill)
        res=df.transpose()
        res.index.name="Product ID"
        prod=[bill[prod_id]['Price']*bill[prod_id]['Quantity'] for prod_id in bill.keys()]
        res['Net Amount']=prod
        display(res)
        total_amt=sum(prod)
        print("\n\nYOUR TOTAL PAYABLE AMOUNT = ", total_amt)
```

# Driver code for the user interface to view, purchase or generate bill

```python
cust_count=1
iter_count=1
while True:
    print("\n\n******ABC MEDICINE HALL******")
    if cust_count==1 and iter_count==1:
        print("\n------WELCOME CUSTOMER 1------\n")
        iter_count=0
    print("1. New customer, reset bill (no need for 1st customer)")
    print("2. Display the medicine catalog")
    print("3. Buy product/s")
    print("4. Print your Bill")
    print("0. Exit")
    print("Enter choice: ")
    ch=input()
    if ch=='1':
        bill.clear()
        cust_count+=1
        print("Bill reset successfull!")
        print("\n------WELCOME CUSTOMER {}------\n".format(cust_count))
    elif ch=='2':
        display_catalog()
    elif ch=='3':
        purchase_prod()
    elif ch=='4':
        print("\n### BILL FOR CUSTOMER {} ###\n".format(cust_count))
        generate_bill()
    else:
        print("\n~~~~THANK YOU FOR YOUR PURCHASES~~~~\n")
        print("\n{} customers visited today".format(cust_count))
        break
```
